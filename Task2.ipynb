```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.patches as mpatches
import sklearn.metrics as sm
from mpl_toolkits.mplot3d import Axes3D
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import DBSCAN
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
print(iris.data)
```

```python
print(iris.target_names)
```

```python
print(iris.target)
```

```python
x = iris.data
y = iris.target
```

```python
fig = plt.figure(1,figsize=(7,5))
ax = Axes3D(fig, rect=[0,0,0.95,1],elev=48,azim=134)
ax.scatter(x[:,3],x[:,0],x[:,2], edgecolor='k',s=50)
ax.set_xlabel('Petal width')
ax.set_ylabel('Sepal length')
ax.set_zlabel('Petal length')
plt.title('Iris Clusering k Means = 3', fontsize=14)
plt.show()
```

```python
hier = linkage(x,'ward')
maxD = 7.08
plt.figure(figsize=(15,8))
plt.title('Iris Hierarchical Clustering Dendograms')
plt.xlabel('Species')
plt.ylabel('Distance')
dendrogram(hier,truncate_mode='lastp',p=50,leaf_rotation=90,leaf_font_size=8,)
plt.axhline(y=maxD,c='k')
plt.show()
```

```python
x = pd.DataFrame(iris.data, columns=['Sepal Length','Sepal Width','Petal Length','Petal Width'])
y = pd.DataFrame(iris.target,columns=['Target'])
x.head()
```

```python
y.head()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
print(kmeans.labels_)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(14,6))
colors = np.array(['blue','yellow','red'])
predY = np.choose(kmeans.labels_,[1,0,2]).astype(np.int64)
```

```python
plt.subplot(1,2,1)
plt.scatter(x['Petal Length'],x['Petal Width'],c=colors[y['Target']])
plt.title("Before classification")
plt.legend()

plt.subplot(1,2,2)
plt.scatter(x['Petal Length'],x['Petal Width'],c=colors[predY])
plt.title("After Model's classification")
plt.legend()
```

```python
sm.accuracy_score(predY,y['Target'])
```

```python
sm.confusion_matrix(predY,y['Target'])
```
